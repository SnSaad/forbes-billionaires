Creating a dictonary for each person with their birht_year

```python
by=dict()
for i in range(len(w)):
    indx = w.get_group(l[i]).index.to_list()
    
    for j in indx:
        if not math.isnan(df.loc[j,'birth_year']):
            by[l[i]]=df.loc[j,'birth_year']
            break
```

There are different birth_year for same person so we are using the above dictonary to make it same in the below code

```python
for i in range(len(w)):
    indx = w.get_group(l[i]).index.to_list()
    
    for j in indx:
        if l[i] in by and df.loc[j,'birth_year']!=by[l[i]]:
            df.loc[j,'birth_year']=by[l[i]]
```

<h3>Handling wealth_status column</h3>

```python
w = df.groupby('full_name')
```

```python
l=[]
for i in df['full_name'].unique():
    l.append(i)
```

```python
len(l)
```

```python
s=0
ctr=0
for i in range(len(w)):
    indx = w.get_group(l[i]).index.to_list()
    d = w.get_group(l[i])[['year','net_worth','wealth_status','full_name']]
            
    if len(indx)==1:
        if type(df.loc[indx[0],'wealth_status'])==float:
            df.loc[indx[0],'wealth_status']='Increased'
        continue
    
    for p,c in zip(indx,indx[1:]):
        ws = df.loc[p,'wealth_status']
        y = int(d.loc[p,['year']]) - int(d.loc[c,['year']])
        n = float(d.loc[p,['net_worth']]) - float(d.loc[c,['net_worth']])
        
            
#         print(y,n,ws)
        if type(ws)==float:
            if y==-1:
                if n==0:
                    df.loc[p,'wealth_status'] = 'Remained Even'
                elif n<0:
                    df.loc[p,'wealth_status'] = 'Increased'
                elif n>0:
                    df.loc[p,'wealth_status'] = 'Decreased'
            elif y<-1:
                df.loc[p,'wealth_status'] = 'Returned to list'
#  list me jo last hai usme change nhi ho rha tha isiliye niche ka code likhna pada
    if type(df.loc[indx[-1],'wealth_status'])==float:
        ws = df.loc[indx[-1],'wealth_status']
        y = int(d.loc[indx[-1],['year']]) - int(d.loc[indx[-2],['year']])
        n = float(d.loc[indx[-1],['net_worth']]) - float(d.loc[indx[-2],['net_worth']])
        
        if type(ws)==float:
            if y==1:
                if n==0:
                    df.loc[indx[-1],'wealth_status'] = 'Remained Even'
                elif n>0:
                    df.loc[indx[-1],'wealth_status'] = 'Increased'
                elif n<0:
                    df.loc[indx[-1],'wealth_status'] = 'Decreased'
            elif y>1:
                df.loc[indx[-1],'wealth_status'] = 'Returned to list'
```

# Is there any person who got in the list and his/her wealth never decreases until he/she got out of the list?

```python
d={}
l=[]
for i in range(len(data)):
    w=data.loc[i,'wealth_status']
    f=data.loc[i,'full_name']
    if f not in d.keys():
        d[f]=w
    if f in d.keys():
        if d[f]!='Increased' and w!='Increased':
            d[f]=w
            if f in l:
                l.remove(f)
            continue
        elif d[f]=='Increased' and w!='Increased':
            d[f]=w
            if f in l:
                l.remove(f)
            continue
        elif d[f]=='Increased' and w=='Increased':
            d[f]=w
            if f not in l:
                l.append(f)
            
```
